```python
#주가 데이타 대시보드
import yfinance as yf
import plotly.graph_objects as go
from dash import Dash, dcc, html

# 1. 주가 데이터 가져오기
samsung = yf.Ticker('005930.KS')  # 삼성전자 코스피 티커
hist = samsung.history(period='1y')  # 1년간의 주가 데이터

# 2. 대시보드 앱 생성
app = Dash(__name__)

app.layout = html.Div(children=[
    html.H1(children='삼성전자 주가 대시보드'),

    # 주가 차트
    dcc.Graph(
        id='stock-chart',
        figure={
            'data': [
                go.Candlestick(
                    x=hist.index,
                    open=hist['Open'],
                    high=hist['High'],
                    low=hist['Low'],
                    close=hist['Close'],
                    name='삼성전자 주가'
                )
            ],
            'layout': go.Layout(
                title='삼성전자 주가 차트',
                xaxis_title='Date',
                yaxis_title='Price (KRW)',
                xaxis_rangeslider_visible=False
            )
        }
    )
])

# 3. 서버 실행
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
##2024년 주가 변화 곡
# 2024년 데이터만 필터링
hist_2024 = hist[hist.index.year == 2024]

# 2024년 주가 변화 확인
print(hist_2024.head())
```

```python
# 2024년 일별 종가 변화 라인 차트 그리기
fig = go.Figure()
fig.add_trace(go.Scatter(x=hist_2024.index, y=hist_2024['Close'],
                         mode='lines+markers', name='2024년 종가'))

# 레이아웃 설정
fig.update_layout(
    title='2024년 삼성전자 주가 변화',
    xaxis_title='Date',
    yaxis_title='Closing Price (KRW)',
    xaxis_rangeslider_visible=True
)

# 차트 출력
fig.show()
```

```python
# 파이차트로 상승과 하락 시각화
import yfinance as yf
import pandas as pd

# 1. 삼성전자 주가 데이터 가져오기
samsung = yf.Ticker('005930.KS')
hist = samsung.history(period='1y')  # 1년간의 데이터

# 2. 상승과 하락 일수 계산
hist['Price Change'] = hist['Close'].diff()  # 종가 변화 계산
rise_count = hist[hist['Price Change'] > 0].shape[0]  # 상승 일수
fall_count = hist[hist['Price Change'] < 0].shape[0]  # 하락 일수

# 상승과 하락 일수 확인
print(f'상승 일수: {rise_count}, 하락 일수: {fall_count}')
```

```python
import plotly.graph_objects as go

# 1. 파이차트 데이터 생성
labels = ['Rise', 'Fall']
values = [rise_count, fall_count]

# 2. 파이차트 그리기
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.3)])
fig.update_layout(title_text='상승과 하락 비율 (1년)')
fig.show()
```

```python
# 1. 월별 주가 변화 계산
hist['Month'] = hist.index.to_period('M')  # 날짜를 월로 변환
monthly_avg = hist.groupby('Month')['Close'].mean()  # 월별 평균 종가

# 2. 월별 주가 변화 시각화
fig = go.Figure()
fig.add_trace(go.Scatter(x=monthly_avg.index.astype(str), y=monthly_avg,
                         mode='lines+markers', name='월별 평균 주가'))
fig.update_layout(title='월별 주가 변화 추이', xaxis_title='Month', yaxis_title='Average Closing Price')
fig.show()
```

```python
# 1. 요일별 상승/하락 데이터 계산
hist['Weekday'] = hist.index.weekday  # 0 = Monday, 6 = Sunday
weekday_rise_fall = hist.groupby('Weekday')['Price Change'].apply(lambda x: (x > 0).sum())  # 상승 일수
weekday_fall = hist.groupby('Weekday')['Price Change'].apply(lambda x: (x < 0).sum())  # 하락 일수

# 요일별 상승/하락 막대그래프 시각화
fig = go.Figure()
fig.add_trace(go.Bar(x=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], y=weekday_rise_fall, name='상승 일수'))
fig.add_trace(go.Bar(x=['Mon', 'Tue', 'Wed', 'Thu', 'Fri'], y=weekday_fall, name='하락 일수'))

fig.update_layout(barmode='group', title='요일별 상승/하락 추이', xaxis_title='Weekday', yaxis_title='Number of Days')
fig.show()
```
