```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
url = 'https://www.paxnet.co.kr/stock/report/report?menuCode=2222&currentPageNo=1&reportId=0&searchKey=stock&searchValue='
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
table_list = soup.find('div', {'class':'board-type'})
table_list
```

```python
li_tag = soup.find('li', class_='board-list-th')
texts = [div.text for div in li_tag.find_all('div')]

columns = texts
print(columns[0])
print(columns[1])
print(columns[2])
print(columns[3])
print(columns[4])
```

```python
df = pd.DataFrame(columns=columns)
df["종목"]
```

```python
종목_list = []
제목_list = []
적정가격_list = []
투자의견_list = []
제공출처_list = []
```

```python
for i in range(1, 757):
    url = f'https://www.paxnet.co.kr/stock/report/report?menuCode=2222&currentPageNo={i}&reportId=0&searchKey=stock&searchValue='
    time.sleep(1)
    # 요청 보내기
    response = requests.get(url)

    # 요청 성공 확인
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table_list = soup.find('div', {'class': 'board-type'})

        # table_list가 None인지 확인
        if table_list:
            for li_tag in table_list.find_all('li')[1:31]:  # 첫 번째 li 태그를 제외하고 30개까지
                div_tags = li_tag.find_all('div')

                # 각 리스트에 텍스트를 순서대로 추가
                if len(div_tags) >= 5:  # div_tags가 5개 이상인지 확인
                    종목_list.append(div_tags[0].text.strip())
                    제목_list.append(div_tags[1].text.strip())
                    적정가격_list.append(div_tags[2].text.replace("적정가격", "").strip())
                    투자의견_list.append(div_tags[3].text.strip())
                    제공출처_list.append(div_tags[4].text.strip())
        else:
            print(f"Warning: 'board-type' div not found on page {i}.")
    else:
        print(f"Error: Unable to access page {i}. Status code: {response.status_code}")
```

```python
df = pd.DataFrame({
    "종목": 종목_list,
    "제목": 제목_list,
    "적정가격": 적정가격_list,
    "투자의견": 투자의견_list,
    "제공출처": 제공출처_list
    })
df.to_csv("paxnet.csv", index=False)
```
